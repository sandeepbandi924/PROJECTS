### Terraforming Mars and Public Opinion

```python
#importing libraries
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline

import warnings
warnings.filterwarnings('ignore')
```

```python
#Reading all files
filenames=['Xvst-q0.0000-N10000-T1000.txt','Xvst-q0.0250-N10000-T1000.txt','Xvst-q0.0500-N10000-T1000.txt',
           'Xvst-q0.0750-N10000-T1000.txt','Xvst-q0.1000-N10000-T1000.txt','Xvst-q0.1250-N10000-T1000.txt',
           'Xvst-q0.1500-N10000-T1000.txt','Xvst-q0.2000-N10000-T1000.txt']

q_values = [0.0001, 0.0025, 0.0050, 0.0075, 0.0100, 0.0125, 0.0150, 0.0175]
```

```python
# Initialize lists to store results
X_avg_list = [] #avg
X_max_list = []  #max
X_min_list = []  #min
d_list = []  #diff
sigma_list = []  #std
```

```python
# Loop each file analyze the data
for filename, q in zip(filenames, q_values):
    # Load the data
    data = np.loadtxt(filename)
    t = data[:, 0]
    X = data[:, 1]
    
    # Plot X vs t
    plt.figure()
    plt.plot(t, X, label=f'q={q}')
    plt.xlabel('Day (t)')
    plt.ylabel('Average Opinion (X)')
    plt.title(f'X vs t for q={q}')
    plt.legend()
    plt.show()
    
    # statistics analysis
    X_avg = np.mean(X[t >= 200])
    X_max = np.max(X[t >= 200])
    X_min = np.min(X[t >= 200])
    d = X_max - X_min
    sigma = np.std(X[t >= 200])
    
    # appending results
    X_avg_list.append(X_avg)
    X_max_list.append(X_max)
    X_min_list.append(X_min)
    d_list.append(d)
    sigma_list.append(sigma)
    
    # Print results for each q
    print(f'q={q}: X_avg={X_avg}, X_max={X_max}, X_min={X_min}, d={d}, sigma={sigma}')
```

```python
# Plot X_avg vs q
plt.figure()
plt.plot(q_values, X_avg_list, marker='o')
plt.xlabel('q')
plt.ylabel('Average X (X_avg)')
plt.title('Average X vs q')
plt.show()
```

```python
# Plot X_max vs q
plt.figure()
plt.plot(q_values, X_max_list, marker='o')
plt.xlabel('q')
plt.ylabel('Maximum X (X_max)')
plt.title('Maximum X vs q')
plt.show()
```

```python
# Plot X_max vs q
plt.figure()
plt.plot(q_values, X_max_list, marker='o')
plt.xlabel('q')
plt.ylabel('Maximum X (X_max)')
plt.title('Maximum X vs q')
plt.show()
```

```python
# Plot d vs q
plt.figure()
plt.plot(q_values, d_list, marker='o')
plt.xlabel('q')
plt.ylabel('Distance (d)')
plt.title('Distance (d) vs q')
plt.show()
```

```python

# Plot sigma vs q
plt.figure()
plt.plot(q_values, sigma_list, marker='o')
plt.xlabel('q')
plt.ylabel('Standard Deviation ()')
plt.title('Standard Deviation () vs q')
plt.show()
```

```python
# Plot histograms for each dataset
for filename, q in zip(filenames, q_values):
    # Load the data
    data = np.loadtxt(filename)
    t = data[:, 0]
    X = data[:, 1]
    
    # Plot histogram
    plt.figure()
    plt.hist(X[t >= 200], bins=30, alpha=0.7, label=f'q={q}')
    plt.xlabel('Average Opinion (X)')
    plt.ylabel('Frequency')
    plt.title(f'Histogram of X for q={q}')
    plt.legend()
    plt.show()
```

```python
# Find the q value with maximum and minimum sigma
max_sigma_q = q_values[np.argmax(sigma_list)]
min_sigma_q = q_values[np.argmin(sigma_list)]

print(f'The value of q that yields the maximum value of  is {max_sigma_q}')
print(f'The value of q that yields the minimum value of  is {min_sigma_q}')
```
